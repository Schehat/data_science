```python
import pandas as pd
```

```python
df = pd.read_csv("survey_results_public.csv", index_col="Respondent")
schema_df = pd.read_csv("survey_results_schema.csv", index_col="Column")
```

* ### sort by 1 column

```python
df.sort_values(by="ConvertedComp")
df.sort_values(by="ConvertedComp", ascending=False)
```

* ### sort by multiple columns

```python
df.sort_values(by=["ConvertedComp", "Country"], ascending=False)  # both descending order
df.sort_values(by=["ConvertedComp", "Country"], ascending=[False, True])  # different sorting order
```

* ### sort_index

```python
df.sort_index()
```

* ### sort_values: applies to series

```python
df["ConvertedComp"].sort_values()
```

* ### series: get nlargest/nsmallest

```python
df["ConvertedComp"].nlargest(10)
```

* ### dataframe: get nlargest/nsmallest

```python
# by taking the nlargest of the series into account
# df.loc[df["ConvertedComp"].nlargest(10).index, :]

# simpler way of doing this
df.nlargest(10, "ConvertedComp")
```
