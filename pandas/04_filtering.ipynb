```python
import pandas as pd
```

```python
df = pd.read_csv("survey_results_public.csv", index_col="Respondent")
schema_df = pd.read_csv("survey_results_schema.csv", index_col="Column")
```

* ### logical operations: and: &, or: |, not: ~
* ### brackets are necessary when chaining logical operations 

* ### filtering with index: cant set columns

```python
df[(df["Hobbyist"] == "Yes") & (df["OpenSourcer"] == "Never")] 
```

* ### filtering with loc: can set columns 

```python
df.loc[~(df["Hobbyist"] == "Yes"), ["MainBranch", "OpenSourcer"]]
```

```python
df.loc[df["ConvertedComp"] > 70000, ["ConvertedComp", "Country", "LanguageWorkedWith"]]
```

* ### isin

```python
# pass list with isin-method
countries = ["United States", "United Kingdom", "India", "Canada", "Germany"]
df.loc[df["Country"].isin(countries), "Country"]
```

* ### use methods on series datatype: str.contains

```python
filt = df["LanguageWorkedWith"].str.contains("Python", na=False)
df.loc[filt, "LanguageWorkedWith"]
```
