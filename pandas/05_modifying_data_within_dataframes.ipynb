```python
import pandas as pd
```

```python
df = pd.read_csv("survey_results_public.csv", index_col="Respondent")
schema_df = pd.read_csv("survey_results_schema.csv", index_col="Column")
```

* ### set column names to lowercase

```python
df.columns = [x.lower() for x in df.columns]
df.head()
```

* ### replace whitespace with underscores

```python
df.columns = df.columns.str.replace(" ", "_")
df.head()
```

* ### renaming columns

```python
# df.rename(columns={"mainbranch" : "mainarea", "hobbyist" : "hobby"}, inplace=True)
# df.head()
```

* ### modify values of a row: loc & at
* ### at is faster than loc when updating 1 value

```python
df.loc[1, ["mainbranch", "hobbyist"]] = ["developer", "no"]

# no list needed for single assignment
df.at[1, "opensourcer"] = "Yes"

df.head()
```

* ### modify entire column

```python
df["hobbyist"] = df["hobbyist"].str.lower()
df.head()
```

## apply

* ### works on dataframe & series
* ### apply expects a function as a parameter. Passing custom functions possible

* ### apply on series

```python
df["hobbyist"].apply(len)  # applies len on all element in the series
```

```python
# example with custom function
df["hobbyist"].apply(lambda x: len(x))
```

* ### apply on dataframe: does not apply to every element in the dataframe, <br>instead to every series in the dataframe

```python
# counts amount of rows for each column
# axis: default=0, 0:columns & 1:rows
df.apply(len, axis="rows")
```

```python
len(df["hobbyist"])  # equivalent instead of apply to whole dataframe
```

```python
df.apply(len, axis="columns")  # counts amount of columns for each row
```

```python
df.apply(str).apply(min)
# df.apply(str).apply(lambda x: min(x))
```

## applymap

* ### only works with dataframes & applies function to every element in the dataframe

```python
df.applymap(str).applymap(len)
```

* ### lowercase every element, na_action is a nice feature

```python
df.applymap(str).applymap(str.lower, na_action="ignore")
```

## map

* ### only work with series
* ### substitutes every element according to the dictionary and the rest with NaN

```python
df["country"].map({"United Kingdom" : "U.K.", "Thailand" : "Thai"})
```

## replace
* ### identical to map but does not replace not included elements in the dictionary with NaN

```python
df["country"].replace({"United Kingdom" : "U.K.", "Thailand" : "Thai"})
```
