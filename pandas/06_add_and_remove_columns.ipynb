```python
import pandas as pd
```

```python
df = pd.read_csv("survey_results_public.csv", index_col="Respondent")
schema_df = pd.read_csv("survey_results_schema.csv", index_col="Column")
```

## columns

* ### create new column

```python
df["new_column1"] = df["MainBranch"] + " --- " + df["Hobbyist"]
df["new_column2"] = df["MainBranch"].apply(str).apply(len)
df.head()
```

* ### drop columns

```python
# 1 columns passing as list not necessary
df.drop(columns=["MainBranch", "Hobbyist"])
```

* ### split series & convert to dataframe

```python
# only splits in a series
df["MainBranch"].str.split(" ")
# converts to dataframe
df["MainBranch"].str.split(" ", expand=True)
```

* ### create multiple columns

```python
df[[x for x in range(16)]] = df["MainBranch"].str.split(" ", expand=True)
df
```
